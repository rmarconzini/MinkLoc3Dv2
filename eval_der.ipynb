```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica i dati
df = pd.read_csv("/Users/marconzinir/Personal/tesi/git/MinkLoc3Dv2/output_eval_der_wandb/dauntless-flower-7/oxford.csv")

# Converte e ordina per % Kept decrescente
df["% Kept"] = df["% Kept"].astype(int)
df = df.sort_values("% Kept", ascending=False)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(df["% Kept"], df["Avg top 1% Recall"], marker='o')

# Imposta le etichette sull'asse X
plt.xticks(df["% Kept"])  # mostra solo i valori presenti nel CSV

plt.gca().invert_xaxis()  # Inverti l'asse X: da 100  0

plt.xlabel("% Kept ()")
plt.ylabel("Avg top 1% Recall")
plt.title("Oxford: Avg top 1% Recall vs % Kept")
plt.grid(True)
plt.tight_layout()
plt.show()
```
